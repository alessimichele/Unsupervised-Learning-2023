```python
import math
import pandas as pd
import numpy as np
from ClusteringClass import Clustering
import random
```

```python
X = np.genfromtxt("../Datasets/s3.txt")
y = np.genfromtxt("../Datasets/s3.txt")
```

```python
random_integers = [random.randint(0, 50) for i in range(20)]
```

```python
losses = np.empty(20)
for i in range(20):
    k=Clustering(k=15, algorithm_variant='kmeans')
    k.fit(X, random_choice=random_integers[i])
    losses[i] = k.loss

minimum, maximum = np.min(losses), np.max(losses)
mean = np.mean(losses)
print(minimum, " ", maximum, " ", mean)
```

```python
losses = np.empty(20)
for i in range(20):
    k=Clustering(k=15, algorithm_variant='kmeans++')
    k.fit(X, random_choice=random_integers[i])
    losses[i] = k.loss

minimum, maximum = np.min(losses), np.max(losses)
mean = np.mean(losses)
print(minimum, " ", maximum, " ", mean)
```

```python
best = np.argmin(losses)
# recover the seed
seed = random_integers[best]

k=Clustering(k=15, algorithm_variant='kmeans++')
k.fit(X, random_choice=seed)
k.scree_test(X, max_k=20)
```
