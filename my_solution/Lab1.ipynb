# Gaussian
Generate data from two random Gaussians, each of which represents a class. 

Flip randomly a percentage of the labels and plot the results.

```python
# generate random mean and standard deviation values
np.random.seed(1)
mu = np.random.randint(0,100,2)
sd = np.random.uniform(0,5,2)

# create two Pandas Series with random values generated from a normal distribution
class1 = pd.Series(np.random.normal(mu[0], sd[0], 100))
class2 = pd.Series(np.random.normal(mu[1], sd[1], 100))

# randomly flip a percentage of the labels
flip_percent = 20 # flip 20% of the labels
num_flips = int((flip_percent/100)*len(class1))
flip_indices = np.random.choice(range(len(class1)), size=num_flips, replace=False)
class1[flip_indices], class2[flip_indices] = class2[flip_indices], class1[flip_indices] #swap

# create a DataFrame with two columns, "Class 1" and "Class 2"
df = pd.DataFrame({"Class 1": class1, "Class 2": class2})

df
```

```python
plt.scatter(class1, class2)
```

## Swiss roll 
It is defined by the following mapping:
$$x = \phi\cos(\phi) , y= \phi\sin(\phi) , z = \psi$$
with $\phi \in(1.5\pi, 4.5\pi)$ and $\psi \in(0,10)$


Create the dataset of $n=1000$ points, plot them in 2d and 3d.

```python
phi = np.linspace(1.5*math.pi, 4.5*math.pi, 1000)

x = phi*np.cos(phi)
y = phi*np.sin(phi)

plt.plot(x,y)
```

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the mapping
def mapping(phi, psi):
    x = phi * np.cos(phi)
    y = phi * np.sin(phi)
    z = psi
    return x, y, z

# Define the range of phi and psi
phi_min, phi_max = 1.5 * np.pi, 4.5 * np.pi
psi_min, psi_max = 0, 10

# Generate random values for phi and psi
phi = np.random.uniform(phi_min, phi_max, size=100000)
psi = np.random.uniform(psi_min, psi_max, size=100000)


print(psi)

# Apply the mapping to the random values of phi and psi
x, y, z = mapping(phi, psi)

# Plot the dataset in 2D
plt.scatter(x, y, s=5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Plot the dataset in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=5)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
print(z)
```
